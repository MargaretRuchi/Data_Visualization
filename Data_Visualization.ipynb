```python
import pandas as pd

# Load the cleaned dataset
df = pd.read_csv('cleaned_titanic.csv')  
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
numerical_df = df.select_dtypes(include=['int64', 'float64'])
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap (Numerical Features)')
plt.tight_layout()
plt.savefig('viz_correlation_heatmap.png')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.catplot(x='Pclass', hue='Sex', col='Survived', data=df, kind='count')
plt.savefig('viz_survival_class_gender.png')
plt.show()
```

```python
df['Sex'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(6,6), title='Gender Distribution')
plt.savefig('viz_gender_distribution.png')
plt.show()
```

```python
sns.kdeplot(data=df[df['Survived']==1]['Age'], label='Survived', fill=True)
sns.kdeplot(data=df[df['Survived']==0]['Age'], label='Did Not Survive', fill=True)
plt.title('Age Distribution by Survival')
plt.legend()
plt.savefig('viz_age_kde_by_survival.png')
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Fare vs Age Colored by Survival')
plt.savefig('viz_fare_vs_age_scatter.png')
plt.show()
```

```python
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.savefig('viz_survival_count.png')
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Sex')
plt.savefig('viz_survival_by_sex.png')
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival by Passenger Class')
plt.savefig('viz_survival_by_class.png')
plt.show()
```

```python
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare Distribution by Survival')
plt.savefig('viz_fare_distribution_survival.png')
plt.show()
```

```python
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title('Survival by Embarked Port')
plt.savefig('viz_survival_by_embarked.png')
plt.show()
```

```python
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived')
plt.savefig('viz_pairplot.png')
plt.show()
```
